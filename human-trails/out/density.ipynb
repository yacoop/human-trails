```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
import numpy as np
x, y, z = np.loadtxt('density.txt', unpack=True)
N = int(len(z)**.5)
r = np.array([max(i, 0.2*1e3) for i in z])
r = r.reshape(N, N)
z = z.reshape(N, N)
plt.imshow(z+1, extent=(np.amin(x), np.amax(x), np.amin(y), np.amax(y)),
        cmap=cm.hot, norm=LogNorm())
plt.colorbar()
plt.show()
print(np.amax(z))
```
